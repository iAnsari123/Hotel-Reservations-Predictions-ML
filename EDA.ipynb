### Dataset Correlation

Converting the object column to categories to avoid correlation error

```python
for i in list(zip(df.dtypes.index,df.dtypes)):
   if i[1] == 'object':
    df[i[0]] = df[i[0]].astype('category')
```

#### checking dtypes again

```python
df.dtypes
```

#### considering only numerial columns

```python
numeric = []
for i in list(zip(df.dtypes.index,df.dtypes)):
   if i[1] == 'int64':
    numeric.append(i[0])
```

### EDA questions

#### What is the distribution of the 'booking_status' variable?


```python
sns.countplot(x='booking_status', data=df)
plt.title('Count of Booking Status')
plt.show()
```

#### How many bookings are for adults only, children only, and a combination of both?

```python
sns.countplot(x='booking_status', data=df,hue='no_of_adults')
plt.title('Count of Each Type of Booking given for adults')
plt.show()
```

```python
sns.countplot(x='booking_status', data=df,hue='no_of_children')
plt.title('Count of Each Type of Booking Status given for Children')
plt.show()
```

#### What is the distribution of the 'type_of_meal_plan' variable?

```python
sns.countplot(x='type_of_meal_plan', data=df)
plt.title('Count of Each Type of Meal Plan')
plt.show()
```

#### How many bookings involve a required car parking space?

```python
plt.figure(figsize=(10,5))
sns.countplot(x='room_type_reserved', data=df, hue='required_car_parking_space')
plt.title('Count of Package required parking space')
plt.show()
```

#### How does the 'market_segment_type' influence the booking status?

```python
sns.countplot(x='market_segment_type', data=df)
plt.title('Count of Each Type of Booking media')
plt.show()
```

#### Average Price per Room for Each Meal Plan?

```python

plt.figure(figsize=(8, 6))
sns.violinplot(x='type_of_meal_plan', y='avg_price_per_room', data=df)
plt.title('Violin Plot of Average Price per Room for Each Meal Plan')
plt.show()
```

#### Number of Children in Each Meal Plan

```python
plt.figure(figsize=(8, 6))
sns.pointplot(x='type_of_meal_plan', y='no_of_children', data=df)
plt.title('Point Plot of Number of Children in Each Meal Plan')
plt.show()
```
